# UnSupervised Learning

## Plot of Kmeans with Elbow method

```python
# Plot of Kmeans with Elbow method
Kmeans_model = KMeans()
elbowMethod = KElbowVisualizer(Kmeans_model, k=(1,10))

y = df_total_imp['CUSTOMER']
X = df_total_imp.drop('CUSTOMER', axis=1)

elbowMethod.fit(X)
elbowMethod.show()
```

```python
# Convert target type for H2O
df_total_imp = df_total_imp.astype(dtype = {'CUSTOMER':'float'})

# Start model h2o
h2o.init()

# Import the dataset into H2O
hf = h2o.H2OFrame(df_total_imp)

# For binary classification, response should be a factor
hf['CUSTOMER']=hf['CUSTOMER'].asfactor()
```

```python
# Split the dataset into a train and valid set
train, valid = hf.split_frame(ratios = [.8], seed = seed)


# Build and train the model
hf_kmeans = H2OKMeansEstimator(k=4,
                                 estimate_k=True,
                                 standardize=False,
                                 seed=seed)
hf_kmeans.train(x=hf.col_names,
                  training_frame=train,
                  validation_frame=valid)
```

```python
# Eval performance:
performance = hf_kmeans.model_performance()
performance
```

```python
predict = hf_kmeans.predict(hf)
```

```python
cluster = predict.as_data_frame()[['predict']]
cluster.columns=['CLUSTER']
hf_cluster = pd.concat([df_total_imp.reset_index()['LNR'], hf['CUSTOMER'].as_data_frame(), cluster], axis=1)
```

```python
hf_cluster.to_csv('hf_cluster_4.csv', index=False)
```

```python
# Distribution graph of population by clusters

plt.figure(figsize=(15,5))

# set width of bar
barWidth = 0.25
label=[0,1,2,3,4,5,6]

# set height of bar
bars1 = calculate_cluster(hf_cluster[hf_cluster['CUSTOMER']==0], 'CLUSTER')['Pop_Perc']
bars2 = calculate_cluster(hf_cluster[hf_cluster['CUSTOMER']==1], 'CLUSTER')['Pop_Perc']

# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Make the plot
plt.bar(r1, bars1, width=barWidth, edgecolor='white')
plt.bar(r2, bars2, width=barWidth, edgecolor='white')
 
# Add xticks on the middle of the group bars
plt.title(('Ratio of Population Azdias & Customer\n segments as (%) of total per Cluster'), size=16)
plt.xlabel('Cluster', size=14)
plt.ylabel('Percentage', size=14)
plt.xticks([r + barWidth/2 for r in range(len(bars1))], label)
 
# Create legend & Show graphic
plt.legend(('Azdias-General Population', 'Customer Population'),fontsize=15)
plt.show()


plt.figure(figsize=(15,5))
# set height of bar
bars3 = calculate_cluster(hf_cluster[hf_cluster['CUSTOMER']==0], 'CLUSTER')['Population']
bars4 = calculate_cluster(hf_cluster[hf_cluster['CUSTOMER']==1], 'CLUSTER')['Population']
 
# Make the plot
plt.bar(r1, bars3, width=barWidth, edgecolor='white')
plt.bar(r2, bars4, width=barWidth, edgecolor='white')
 
# Add xticks on the middle of the group bars
plt.title(('Ratio of Population Azdias & Customer\n segments as Count of total per Cluster'), size=16)
plt.xlabel('Cluster', size=14)
plt.ylabel('Count', size=14)
plt.xticks([r + barWidth/2 for r in range(len(bars3))], label)
 
# Create legend & Show graphic
plt.legend(('Azdias-General Population', 'Customer Population'),fontsize=15);
```
